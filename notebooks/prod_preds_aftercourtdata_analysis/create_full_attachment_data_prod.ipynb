```python
def flatten_attachment_pred_df(df: pd.DataFrame) -> pd.DataFrame:
    metadata_cols = ['attachment_id', 'document_s3_key', 'document_s3_bucket',
                     'file_name', 'textract_job_id', 'textract_status', 'textract_s3_link', 'text']
    try:
        metadata = df[metadata_cols].iloc[[0]].reset_index(drop=True)
    except Exception:
        print("attachment_id:", df['attachment_id'].iloc[0], "is missing some metadata columns.")
        metadata = pd.DataFrame(columns=metadata_cols)

    preds = df[['attachment_id', 'type', 'subtype', 'value']].copy()
    preds['col_name'] = preds['type'] + '__' + preds['subtype'].fillna('')
    pivoted = preds.pivot_table(
        index='attachment_id', columns='col_name', values='value', aggfunc='first'
    ).reset_index(drop=True)

    return pd.concat([metadata, pivoted], axis=1)
```

```python
import sys
import os
import glob

import boto3
sys.path.append('../../')

from utils.prod_utils import get_data_by_attachment_id

session = boto3.Session(profile_name='739275445236_DataScienceUser')
s3 = session.client('s3')
unique_attchment_ids = data.attachment_id.unique()

CHECKPOINT_DIR = '/Users/melih.gorgulu/Desktop/Projects/aftercourt_automation/notebooks/analysis_data/egvp_full_attachment_data_checkpoints'
CHECKPOINT_PREFIX = 'attachment_data_chunk'
CHECKPOINT_INTERVAL = 1000

os.makedirs(CHECKPOINT_DIR, exist_ok=True)

# Load already processed attachment_ids from existing checkpoints
existing_files = sorted(glob.glob(f'{CHECKPOINT_DIR}/{CHECKPOINT_PREFIX}_*.csv'))
if existing_files:
    already_processed_df = pd.concat([pd.read_csv(f) for f in existing_files], ignore_index=True)
    already_processed_ids = set(already_processed_df['attachment_id'].astype(str).unique())
    print(f"Resuming: {len(already_processed_ids)} attachment_ids already processed across {len(existing_files)} checkpoint(s).")
else:
    already_processed_ids = set()
    print("Starting fresh.")

pending_ids = [aid for aid in unique_attchment_ids if str(aid) not in already_processed_ids]
print(f"Remaining: {len(pending_ids)} attachment_ids to process.")

chunk = []
chunk_start_index = len(already_processed_ids)
n = len(unique_attchment_ids)

for i, attachment_id in enumerate(pending_ids):
    cur_attch_data = get_data_by_attachment_id(attachment_id, analytics_db, s3, verbose=False)
    chunk.append(cur_attch_data)
    processed_count = chunk_start_index + i + 1
    print(f"Processed {processed_count}/{n} attachment_ids", end='\r')

    if len(chunk) >= CHECKPOINT_INTERVAL:
        chunk_df = pd.concat(chunk, ignore_index=True)
        chunk_index = processed_count // CHECKPOINT_INTERVAL
        chunk_path = f'{CHECKPOINT_DIR}/{CHECKPOINT_PREFIX}_{chunk_index:04d}.csv'
        chunk_df.to_csv(chunk_path, index=False)
        print(f"\nSaved checkpoint: {chunk_path}")
        chunk = []

# Save any remaining records
if chunk:
    chunk_df = pd.concat(chunk, ignore_index=True)
    chunk_index = (chunk_start_index + len(pending_ids)) // CHECKPOINT_INTERVAL + 1
    chunk_path = f'{CHECKPOINT_DIR}/{CHECKPOINT_PREFIX}_{chunk_index:04d}_final.csv'
    chunk_df.to_csv(chunk_path, index=False)
    print(f"\nSaved final checkpoint: {chunk_path}")

# Load all checkpoints into a single DataFrame
all_files = sorted(glob.glob(f'{CHECKPOINT_DIR}/{CHECKPOINT_PREFIX}_*.csv'))
attachment_data = pd.concat([pd.read_csv(f) for f in all_files], ignore_index=True)
print(f"\nDone. Total records loaded: {len(attachment_data)}")
```

```python
flattened_dfs = []
for attachment_id, group_df in attachment_data.groupby('attachment_id'):
    try:
        flattened_dfs.append(flatten_attachment_pred_df(group_df))
    except Exception as e:
        print(f"Error processing attachment_id {attachment_id}: {e}")

flattened_df = pd.concat(flattened_dfs, ignore_index=True)
print(f"Flattened DataFrame shape: {flattened_df.shape}")
flattened_df
```
