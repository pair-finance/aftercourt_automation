```python
add_new_data = data.drop(['Ticket uuid', 'egvp_id','Current Ladung Prob', 'Comment'], axis=1)
```

```python
# add text data from current_raw where possible, fetch remaining from DB
text_map = current_raw.set_index('attachment_id')['text'].to_dict()

texts = []
for a_id in add_new_data['attachment_id'].values:
    if a_id in text_map:
        texts.append(text_map[a_id])
    else:
        try:
            result = get_data_by_attachment_id(a_id, analytics_db, s3, verbose=False)
            texts.append(result['text'].iloc[0] if not result.empty and 'text' in result.columns else None)
        except Exception as e:
            print(f"Failed to fetch text for {a_id}: {e}")
            texts.append(None)

add_new_data['text'] = texts
print(f"Total rows: {len(add_new_data)}, Rows with text: {add_new_data['text'].notna().sum()}")
```
